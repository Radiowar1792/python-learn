```python
saisie = input("Veuillez entrer une liste de nombres séparés par des virgules : ")
nombres_str = saisie.split(',')
nombres = [int(nombre.strip()) for nombre in nombres_str]
nombres
# Calcul de la somme avec for et range
somme = 0
for i in range(len(nombres)):
    somme = somme + nombres[i]

print("Somme des nombres :", somme)

somme = sum(nombres)
print("Somme des nombres :", somme)

moyenne = somme / len(nombres)
print("Moyenne des nombres :", moyenne)
a=0
for i in range(len(nombres)):
  if nombres[i]>moyenne:
    a += 1
print(f"Nombres de nombre supérieurs à la moyenne : {a}")
```
